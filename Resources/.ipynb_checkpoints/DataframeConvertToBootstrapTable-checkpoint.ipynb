```python
#import settings
import pandas as pd
```

```python
# Set filepath
filepath = "cities.csv"

# Read csv file into Pandas Dataframe
cities_df = pd.read_csv(filepath, encoding="utf-8")
cities_df.head()
```

```python
cities_df.reset_index(inplace=True)
del cities_df['Unnamed: 0']
del cities_df['index']
cities_df.head()
```

```python
# render dataframe as html
html = cities_df.to_html(classes=["table-striped", "table-hover"])

print(html)
```
